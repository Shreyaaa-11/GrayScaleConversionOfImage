```python
#converting a colored image to gray scale image.

#import the image library. PIL is a library which is mostly used for reading and writing.
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


#converting the image to form of 3D array consisting of three values namely: R G B
org_image = Image.open("cat.jfif")
org_np = np.array(org_image)
print(org_np.shape)
print(f"{org_np[:5,:5]}") #used to obtain the first five pixels

#IN OUTPUT, HEIGHT, WIDTH AND HOW MANY ARRAYS IT CONSISTS OF WILL BE DISPLAYED.
```

```python
gray_np = np.mean(org_np,axis = 2)
print(gray_np.shape)
gray_np_final = gray_np.astype(np.uint8)

#Knowing the shape and dtype of the image as a gray scale image must have only two values and uint dtype.
print(gray_np_final.shape)
print(gray_np_final.dtype)
print(f"{gray_np_final[:3,:3]}")
```

```python
#mulitplying the decimal values with the RGB colour values.
gray2_np = np.dot(org_np[...,:3], [0.2989, 0.5870, 0.1140])

 #as we know the image ranges from 0 to 255 consisting of only unsigned integers, we must convert our obtained decimals to unsigned values too.
gray2_np_final = gray2_np.astype(np.uint8)
print(f"{gray2_np_final[:3,:3]}")
```

```python
#Converting as image from the array.
gray_image = Image.fromarray(gray_np_final)
plt.imshow(gray_np_final, cmap='gray')
plt.axis('off')  # Hide axes for a cleaner view
plt.show()
```

```python
gray2_image = Image.fromarray(gray2_np_final)
plt.imshow(gray2_np_final, cmap='gray')
plt.axis('off')  # Hide axes for a cleaner view
plt.show()
```

```python
#creates a negative image where we have inverted the colours.
neg_gray_2_np = 255 - gray_2_np_final
neggray_image = Image.fromarray(neg_gray_2_np)
plt.imshow(neg_gray_2_np, cmap='gray')
plt.axis('off')  # Hide axes for a cleaner view
plt.show()
```
